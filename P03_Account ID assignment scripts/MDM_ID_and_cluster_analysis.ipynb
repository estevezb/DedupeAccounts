```python
import pandas as pd
import os

# Assuming your data is in a CSV file named 'data.csv'
# with 'MDM ID' in column A and 'Cluster ID' in column B
file_path = r'C:\Users\beste\OneDrive - Qral Group\01 Narcan\12 Ops\20231223 - Chargeback Adhoc\02 Internal Controls\Processed_inputs\Dedupped_Results\MDM_ID_assigned_Dedupped_Results__20231227_142658.csv'  # Update with the actual path to your file
data = pd.read_csv(file_path)

#Get the directory of the file
directory= os.path.dirname(file_path)

#Create a new directory for the output
output_file = os.path.join(directory, 'Cluster_Analysis_data.csv')

# Group by 'Cluster ID' and count unique 'MDM IDs' for each cluster
cluster_groups = data.groupby('Cluster ID')['MDM ID'].nunique()

# Create a new column 'Label' and set to 0 (default)
data['Label'] = 0

# Loop through each cluster and assign 1 if multiple unique MDM IDs are found
for cluster_id, unique_count in cluster_groups.items():
    if unique_count > 1:
        data.loc[data['Cluster ID'] == cluster_id, 'Label'] = 1

# Output the result to a new CSV file or directly view it
data.to_csv(output_file, index=False)  # Save the result to a new file
print(data)  # Or print the result to the console
```
